```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_size = 100000
population = np.random.exponential(scale=2, size=population_size)

plt.figure(figsize=(8, 5))
sns.histplot(population, bins=50, kde=True, color='blue')
plt.title('Population Distribution (Exponential)', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
def calculate_sample_means(population, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    return sample_means
```

```python
sample_size = 30
num_samples = 1000

sample_means = calculate_sample_means(population, sample_size, num_samples)

plt.figure(figsize=(8, 5))
sns.histplot(sample_means, bins=30, kde=True, color='green')
plt.title(f'Sampling Distribution (Sample Size = {sample_size})', fontsize=14)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
sample_sizes = [10, 30, 50, 100]

plt.figure(figsize=(14, 10))

for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = calculate_sample_means(population, sample_size, num_samples)

    plt.subplot(2, 2, i)
    sns.histplot(sample_means, bins=30, kde=True, color='purple')
    plt.title(f'Sampling Distribution (Sample Size = {sample_size})', fontsize=12)
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
